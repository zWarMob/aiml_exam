# Exploration of CV dataset

This notebook serves as an example on how to iterate over the CV dataset to create a common group understanding. Furthermore, it explores the structure of the dataset. 

```python
# Import pandas
import pandas as pd
```

```python
# Load the CSV file
df = pd.read_csv('data/resume.csv')
```

```python
# Preview the data, columns and data types:
print(df.head())  # Show first 5 rows
```

```python
for index, row in df.iterrows():
    category = row['Category'] # get the value of the category for the row we iterate
    
    if pd.isnull(category): # check if the category is null
        print(f"Category is missing.")
        continue  # Skip to the next row (if you don't want to iterate rows with no category)
    
    print(f"Category {index}: {category}")
```
